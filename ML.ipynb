```python
import pandas as pd
df = pd.read_csv("Highway-Rail_Grade_Crossing_Accident_Data.csv")
df.shape   #return (rows,cols)
df = df.dropna(subset=['Temperature', 'Visibility', 'Weather Condition',
                 'Number of Cars', 'Driver Condition', 'Driver Condition Code'])

df
```

```python
df.describe() #Shift+Tab to see the function's description
df.values
```

```python
temperature = df[['Temperature']]
print(temperature.shape)
print(temperature.dtypes)
```

```python
visibility = df[['Visibility', 'Visibility Code']]
visibility
```

```python
weather = df[['Weather Condition', 'Weather Condition Code']]
weather
```

```python
num_cars = df[['Number of Cars', 'Driver Condition', 'Driver Condition Code']]
driver = df['Driver Condition']
```

```python
from sklearn.tree import DecisionTreeClassifier
temperature_driver = DecisionTreeClassifier()
temperature_driver.fit(temperature, driver)
predictions = temperature_driver.predict([[70.0], [90.0]])
predictions
```

```python
from sklearn.model_selection import train_test_split
temp_train, temp_test, driver_train, driver_test = train_test_split(temperature, driver, test_size=0.2)

temperature_driver = DecisionTreeClassifier()
temperature_driver.fit(temp_train, driver_train)
predictions = temperature_driver.predict(temp_test)

from sklearn.metrics import accuracy_score
score = accuracy_score(driver_test, predictions)
score
```

```python
import joblib
joblib.dump(temperature_driver, 'temperature_predict.joblib')
```

```python
model = joblib.load('temperature_predict.joblib')
predictions = model.predict([[70.0]])
predictions
```

```python
from sklearn import tree
tree.export_graphviz(temperature_driver, out_file='temp-prediction.dot',
                    feature_names=['temperature'],
                    class_names=sorted(driver.unique()),
                    label='all', rounded=True, filled=True)
```
